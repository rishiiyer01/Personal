```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.array([0, 30, 45, 60, 90, 120, 135, 150, 180, 210, 225, 240, 270, 300, 315, 330])
theta_rad = np.radians(theta)  # Convert degrees to radians

xcirc = np.cos(theta_rad)
ycirc = np.sin(theta_rad)
radj=[0.05+0.1*np.cos(np.pi/4),0.1,0.05+0.05/np.sin(np.pi/4),0.1,0.05+0.1*np.cos(np.pi/4),0.1,0.05+0.05/np.sin(np.pi/4),0.1,0.05+0.1*np.cos(np.pi/4),0.1,0.05+0.05/np.sin(np.pi/4),0.1,0.05+0.1*np.cos(np.pi/4),0.1,0.05+0.05/np.sin(np.pi/4),0.1]
xoct=radj*xcirc
yoct=radj*ycirc
plt.plot(xoct, yoct, '-')  # 'o' specifies that we want to plot points
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Points on a Unit Circle')
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')  # To make x and y axes equal
plt.show()
```
